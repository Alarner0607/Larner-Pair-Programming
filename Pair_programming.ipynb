```python
#Function for value conversion

def conversion():
    feet = int(input("Feet: "))
    inches = int(input("Inches: "))
    meters = (0.3048 * feet) + (0.0254 * inches)
    print(f"{meters} Meters")

#Enter values as only numbers

conversion()
```

```python
#Test 1: All 0s

def conversion():
    feet = int(0)
    inches = int(0)
    meters = (0.3048 * feet) + (0.0254 * inches)
    print(f"{meters} Meters")

    if(meters == 0):
        print("Test 1: All 0s - passed")
    else:
        print("Test 1: All 0s - failed")
    

#Enter values as only numbers

conversion()
```

```python
#Test 2: Negative Numbers

def conversion():
    feet = int(-1)
    inches = int(-3)
    meters = (0.3048 * feet) + (0.0254 * inches)
    print(f"{meters} Meters")

    if((meters >= 0.380 and meters <= 0.382) or (meters <= -0.380 and meters >= -0.382)): # should be +/- 0.381 with leniency of +/- 0.001 given, both positive and negative final values accepted
        print("Test 2: Negative Numbers - passed")
    else:
        print("Test 2: Negative Numbers - failed")
    

#Enter values as only numbers

conversion()
```
